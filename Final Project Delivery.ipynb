```python
# Glassdoor API params
employer_info = {}
failed_page = []
api_partner_id = '27144'
api_key = 'hZ0RCoEqtHk'
action = 'jobs-stats'
user_ip = '10.0.1.5'
```

```python
# Function to obtain the jobs data from Glassdoor. 
# state_keys: contains the full names of all the US States
# category_ids: since the API uses param "jc" to specify the "job category" on their websites. This is the list to specify which category we would like to scrape
# sleep_intv: sleep interval between each API call
# return: A data frame that contains the following data for each US state in a specific state:
#           numJobs: total number of jobs in this state
#           num3star: total number of jobs that has a rating greater than 3
#           num4star: total number of jobs that has a rating greater than 4
#           longitude, latitude: position of the state
def create_job_stats_dataframe(state_keys, category_ids, sleep_intv=0):
    state_scores = {key:0 for key in state_keys}
    state_numjobs = dict(state_scores)
    state_num3stars = dict(state_scores)
    state_num4stars = dict(state_scores)
    state_latitude = dict(state_scores)
    state_longitude = dict(state_scores)
    for job_id in category_ids:
        print "Searching num jobs for category " + job_id
        url_jobstats = "http://api.glassdoor.com/api/api.htm?v=1&format=json&t.p="+api_partner_id+"&t.k="+api_key+"&action="+action+"&userip="+user_ip+"&useragent="+"&jc="+job_id+"&returnStates=true"
        jobstats_url = urllib2.urlopen(url_jobstats).read()
        jobstats_json = json.loads(jobstats_url)
        time.sleep(sleep_intv)
        for state_key in jobstats_json['response']['states'].keys():
            state_numjobs[state_key] += jobstats_json['response']['states'][state_key]['numJobs']
            state_scores[state_key] += jobstats_json['response']['states'][state_key]['score']
            state_latitude[state_key] = jobstats_json['response']['states'][state_key]['latitude']
            state_longitude[state_key] = jobstats_json['response']['states'][state_key]['longitude']
        
        print "Searching num jobs greater than 3 stars for category " + job_id
        url_jobstats = "http://api.glassdoor.com/api/api.htm?v=1&format=json&t.p="+api_partner_id+"&t.k="+api_key+"&action="+action+"&userip="+user_ip+"&useragent="+"&jc="+job_id+"&returnStates=true&minRating=3"
        jobstats_url = urllib2.urlopen(url_jobstats).read()
        jobstats_json = json.loads(jobstats_url)
        time.sleep(sleep_intv)
        for state_key in jobstats_json['response']['states'].keys():
            state_num3stars[state_key] += jobstats_json['response']['states'][state_key]['numJobs']
        
        print "Searching num jobs greater than 4 stars for category " + job_id
        url_jobstats = "http://api.glassdoor.com/api/api.htm?v=1&format=json&t.p="+api_partner_id+"&t.k="+api_key+"&action="+action+"&userip="+user_ip+"&useragent="+"&jc="+job_id+"&returnStates=true&minRating=4"
        jobstats_url = urllib2.urlopen(url_jobstats).read()
        jobstats_json = json.loads(jobstats_url)
        time.sleep(sleep_intv)
        for state_key in jobstats_json['response']['states'].keys():
            state_num4stars[state_key] += jobstats_json['response']['states'][state_key]['numJobs']
            
    num_category = len(category_ids)
    for state_key in state_keys:
        state_scores[state_key] /= num_category
        
    out_df = pd.DataFrame({"score":state_scores, "latitude":state_latitude, "longitude":state_longitude, "numJobs": state_numjobs, "num3stars": state_num3stars, "num4stars": state_num4stars})
    out_df.reset_index(inplace=True)
    out_df.rename(columns={"index":"state"}, inplace=True)
    return out_df
```

## Classification without a Model ##

Before going into statistical models, we would like to see if we can simply use the mean or median of culture, leadership, compensation, opportunity, and balance scores of the company to predict the overall satisfaction. 

```python
company_mean = [np.mean(X[i,0:4]) for i in range(len(X))]
Y_hat_mean = np.array([int(round(x)) for x in company_mean])
company_median = [np.median(X[i,0:4]) for i in range(len(X))]
Y_hat_median = np.array([int(round(x)) for x in company_median])
ac_mean = 0.0
ac_median = 0.0
for i in range(len(X)):
    if Y_hat_mean[i] == Y[i]:
        ac_mean += 1.0
    if Y_hat_median[i] == Y[i]:
        ac_median += 1.0
print "Using the means of the five features, we get accuracy: ", ac_mean/len(X)
print "Using the medians of the five features, we get accuracy: ", ac_median/len(X)
```

The accuracy is not bad. However, we may wonder if we can do better. In addition, we also want to discover some other important information such as which feature is more important.

## Multinomial Logistic Regression ##

We start our analysis with a multinomial logistic regression. By using culture, management, compensation, opportunity, balance, company size, and ceo approval rate as features, we classify the company into five groups based on their ratings (0.5-1.5, 1.5-2.5, 2.5-3.5, 3.5-4.5, 4.5-5.5). Companies are subsetted to those with 10 or more people rating on Glassdoor. Since python does not support ordered logit, we use categorical logistic regression instead, where we can find the accuracy of the prediction increases as the inverse regularization strength C increases. However, it does not go above 0.81 in our graph, which uses 10-fold cross validation to test the accuracy.

```python
C = [0.1, 0.5, 1, 5, 10, 25, 50, 100]
scores = []
for c in C:
    fit_logit = LogisticRegression(C = c).fit(X,Y)
    score_logit=cross_val_score(fit_logit,X,Y,cv=10)
    scores.append(np.mean(score_logit))
plt.plot(C, scores)
plt.title('Inverse of Regularization Strength VS Accuracy')
plt.xlabel('Inverse of Regularization Strength')
plt.ylabel('Accuracy')
plt.show()
```

We try Gaussian Naive Bayes Classifier next. ## Ensemble Learning Classifiers ##

```python
maxn = 30
x_list = range(1,maxn)
rfc_scores = []
for n in xrange(1,maxn):
    RFC = RandomForestClassifier(n_estimators=n).fit(X,Y)
    score_rfc=cross_val_score(RFC,X,Y,cv=10)
    rfc_scores.append(np.mean(score_rfc))
etc_scores = []
for n in xrange(1,maxn):
    ETC = ExtraTreesClassifier(n_estimators=n).fit(X,Y)
    score_etc=cross_val_score(ETC,X,Y,cv=10)
    etc_scores.append(np.mean(score_etc))
abc_scores = []
for n in xrange(1,maxn):
    ABC = AdaBoostClassifier(n_estimators=n).fit(X,Y)
    score_abc=cross_val_score(ABC,X,Y,cv=10)
    abc_scores.append(np.mean(score_abc))
gbc_scores = []
for n in xrange(1,maxn):
    GBC = GradientBoostingClassifier(n_estimators=n,learning_rate=1.0,max_depth=1).fit(X,Y)
    score_gbc=cross_val_score(GBC,X,Y,cv=10)
    gbc_scores.append(np.mean(score_gbc))

a1, = plt.plot(x_list,rfc_scores)
a2, = plt.plot(x_list,etc_scores)
a3, = plt.plot(x_list,abc_scores)
a4, = plt.plot(x_list,gbc_scores)
plt.legend([a1,a2,a3,a4],['Random Forest','Extra Trees','Ada Boost','Gradient Boost'],loc=5)
plt.title("Accuracy of Ensemble Learning Methods")
plt.xlabel("n estimates")
plt.ylabel("Accuracy")
plt.show()
```

Based on the accuracy scores using 10-fold cross validation method, random forest and extra tree methods perform best. We decide to stick to random forest.

```python
import warnings
import pylab as pl
import numpy as np
from scipy import linalg
from sklearn.linear_model import (RandomizedLasso, lasso_stability_path, LassoLarsCV)
from sklearn.feature_selection import f_regression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import auc, precision_recall_curve
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.utils.extmath import pinvh
pl.rcParams['figure.figsize'] = (10.0, 8.0)

def mutual_incoherence(X_relevant, X_irelevant):
    """Mutual incoherence, as defined by formula (26a) of [Wainwright2006].
    """
    projector = np.dot(np.dot(X_irelevant.T, X_relevant),
                       pinvh(np.dot(X_relevant.T, X_relevant)))
    return np.max(np.abs(projector).sum(axis=1))


for conditioning in (1, 1e-4):
    ###########################################################################
    # Simulate regression data with a correlated design
    n_features = 500
    n_relevant_features = 8
    noise_level = .2
    coef_min = .2
    # The Donoho-Tanner phase transition is around n_samples=25: below we
    # will completely fail to recover in the well-conditioned case
    n_samples = 25
    block_size = n_relevant_features

    rng = np.random.RandomState(42)

    # The coefficients of our model
    coef = np.zeros(n_features)
    coef[:n_relevant_features] = coef_min + rng.rand(n_relevant_features)

    # The correlation of our design: variables correlated by blocs of 3
    corr = np.zeros((n_features, n_features))
    for i in range(0, n_features, block_size):
        corr[i:i + block_size, i:i + block_size] = 1 - conditioning
    corr.flat[::n_features + 1] = 1
    corr = linalg.cholesky(corr)

    # Our design
    X = rng.normal(size=(n_samples, n_features))
    X = np.dot(X, corr)
    # Keep [Wainwright2006] (26c) constant
    X[:n_relevant_features] /= np.abs(
        linalg.svdvals(X[:n_relevant_features])).max()
    X = StandardScaler().fit_transform(X.copy())

    # The output variable
    y = np.dot(X, coef)
    y /= np.std(y)
    # We scale the added noise as a function of the average correlation
    # between the design and the output variable
    y += noise_level * rng.normal(size=n_samples)
    mi = mutual_incoherence(X[:, :n_relevant_features],
                            X[:, n_relevant_features:])

    ###########################################################################
    # Plot stability selection path, using a high eps for early stopping
    # of the path, to save computation time
    alpha_grid, scores_path = lasso_stability_path(X, y, random_state=42,
                                                   eps=0.05)

    pl.figure()
    pl.rcParams['figure.figsize'] = (10.0, 8.0)
    # We plot the path as a function of alpha/alpha_max to the power 1/3: the
    # power 1/3 scales the path less brutally than the log, and enables to
    # see the progression along the path
    hg = pl.plot(alpha_grid[1:] ** .333, scores_path[coef != 0].T[1:], 'r')
    hb = pl.plot(alpha_grid[1:] ** .333, scores_path[coef == 0].T[1:], 'k')
    ymin, ymax = pl.ylim()
    pl.xlabel(r'$(\alpha / \alpha_{max})^{1/3}$')
    pl.ylabel('Stability score: proportion of times selected')
    pl.title('Stability Scores Path - Mutual incoherence: %.1f' % mi)
    pl.axis('tight')
    pl.legend((hg[0], hb[0]), ('relevant features', 'irrelevant features'),
              loc='best')

    ###########################################################################
    # Plot the estimated stability scores for a given alpha

    # Use 6-fold cross-validation rather than the default 3-fold: it leads to
    # a better choice of alpha:
    # Stop the user warnings outputs- they are not necessary for the example
    # as it is specifically set up to be challenging.
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', UserWarning)
        lars_cv = LassoLarsCV(cv=6).fit(X, y)

    # Run the RandomizedLasso: we use a paths going down to .1*alpha_max
    # to avoid exploring the regime in which very noisy variables enter
    # the model
    alphas = np.linspace(lars_cv.alphas_[0], .1 * lars_cv.alphas_[0], 6)
    clf = RandomizedLasso(alpha=alphas, random_state=42).fit(X, y)
    trees = ExtraTreesRegressor(100).fit(X, y)
    # Compare with F-score
    F, _ = f_regression(X, y)

    pl.figure()
    pl.rcParams['figure.figsize'] = (10.0, 8.0)
    for name, score in [('F-test', F),
                        ('Stability selection', clf.scores_),
                        ('Lasso coefs', np.abs(lars_cv.coef_)),
                        ('Trees', trees.feature_importances_),
                        ]:
        precision, recall, thresholds = precision_recall_curve(coef != 0,
                                                               score)
        pl.semilogy(np.maximum(score / np.max(score), 1e-4),
                    label="%s. AUC: %.3f" % (name, auc(recall, precision)))

    pl.plot(np.where(coef != 0)[0], [2e-4] * n_relevant_features, 'mo',
            label="Ground truth")
    pl.xlabel("Features")
    pl.ylabel("Score")
    # Plot only the 100 first coefficients
    pl.xlim(0, 100)
    pl.legend(loc='best')
    pl.title('Feature selection scores - Mutual incoherence: %.1f'
             % mi)

pl.show()
```

## Text Mining ##

As the employee’s feedback comments (in natural language form) could also reveal how the company is like and what are the pros and cons of a specific company, we decide to do text mining to dig out how companies are similar to each other with respect to the employee’s feedback comments. 
We index all the files by using the python nltk(Natural Language Toolkit)library. Firstly, we remove all the punctuations and stop words in the comments. Then, we tokenize the comments into words and stem them into their simplest forms. Then, the words will be stored into a dictionary with their total collection frequency and a unique ID to their respective entry in a posting list. The posting list contains the word ID and a list of documents ID that word is contained in and the document frequency of that word. In indexing step, we get the word dictionary and posting file.
####Indexing the pros and cons comments for all the companies

```python
#all files name in the pros and cons
docID_ = []
import re
import string
import nltk
import sys
import getopt
import os
import heapq
import math
import glob
import copy
from nltk.corpus import stopwords
from nltk.stem.porter import *

#for general directory path
dir_path_final = ""

#name by default for dic and postings
file_dic = "dic"       
file_posting = "posting"

#general list to store words
word_list = {}

#doc ID list 
docID_list = {}

#stop words
stop = stopwords.words('english')

#process a file
def process_file(file_name):

    
    try:
        target = open(file_name)
    except:
        return []
    
    content = target.read()

    #replace the "\n" " +" by space
    
    content = re.sub('\n', '', content)
    content= re.sub(r'[^\w]', ' ', content)
    
    target.close()
    file_words = {}

    #tokenize and stem the words in file
    stemmer = PorterStemmer()
    
    for sent_token in nltk.sent_tokenize(content):
        for word in nltk.word_tokenize(sent_token):
            stemmed_word = stemmer.stem_word(word.lower())
            if(stemmed_word not in stop):
                # count the term freq  
                if file_words.has_key(stemmed_word):
                    file_words[stemmed_word] += 1
                else:
                    file_words[stemmed_word] = 1

    result = []    
    
    for key in sorted(file_words.keys()):
        result.append([key,file_words[key]])
        
    #mark the word in the general list
    for word in file_words:
        word_list[word] = 1
    #delete the empty entry
    if word_list.has_key(''):
        del word_list['']
           
    return result

#function to index the documents
def index_doc():
    
    global docID_list
    global docID_
    file_name = []
    
    #sort the file in directory for reading, idea got from stackoverflow
    #http://stackoverflow.com/questions/12093940/reading-files-in-a-particular-order-in-python
    files_sorted = sorted(glob.glob(dir_path_final+'*'),key = lambda a: str(os.path.basename(a)))

    #write temp results to a file
    temp_result = open("temp_result.txt",'w')
    for file_name in files_sorted:
        
        if file_name.isdigit():
            continue
        #get the docID from file
        docID = str(os.path.basename(file_name))
        docID_list[docID] = 1

        #get the docID-(term1, tf)(term2, tf)...
        temp_list = process_file(file_name)
        if len(temp_list)==0:
            continue

        #write the result to temp file
        temp_result.write(docID+' '+str(temp_list)+'\n')
        
    temp_result.close()
    
    #sort doclist and wordlist
    file_name = [x for x in docID_list.keys()]
    file_name.sort()
    docID_.append(file_name)
    #print docID_list

    #use the word_list globally
    global word_list
    word_list_ = sorted(word_list.keys())

    dic_file = open(file_dic,'w')
    post_file = open(file_posting,'w')

    #the total number of documents
    post_file.write(str(len(file_name))+"\n")
    

    temp_dic = {}
    temp_result = open("temp_result.txt",'r')

    #reverse the stucture to term (docID1 tf1) (docID2 tf2)...
    for line in temp_result:
        array = line.split()

        #get and remove the docID
        doc_id = array.pop(0)
            
        if doc_id[0].isdigit():
            continue
        
        #print array, doc_id
        try:
            term_tf_list = eval(''.join(array))
        except:
            continue
            
        #print term_tf_list
        
        for term, tf in term_tf_list:
            
            if temp_dic.has_key(term):
                temp_dic[term].append([doc_id, tf])
            else:
                temp_dic[term]= [[doc_id, tf]]

    temp_result.close()
    #os.remove("temp_result.txt")

    line_pointer = 2
    
    for word in word_list_:
        #print word
        #construct the dic: term df posting_pointer
        
        try:
            dic_file.write(word+" "+str(len(temp_dic[word]))+" "+str(line_pointer)+"\n")
            #construct the posting: [doc_id,tf]...
            post_file.write(str(temp_dic[word])+"\n")
        except:
            continue
            
        #increase the pointer
        line_pointer += 1

    post_file.close()
    dic_file.close()

    
#read in the 
dir_ = ["Comments_pros/", "Comments_cons/"] 

for ele in dir_:
    dir_path_final = ele
    file_dic = "dic"+ele.split("/")[0]
    file_posting = "posting"+ele.split("/")[0]
    index_doc()
```


For each of the company, we select the top 50 companies having the largest similarity. We define similarity by computing the normalized tf-idf score for a specific company pros/cons part of the comment with all the rest company’s pros/cons part of the comment. In this way, we could figure out, from the perspective of the content of comments, how companies could be related and grouped. Then, we plot the cons similarity as the x and pros similarity as y on the graph.
####Functions to compute the similarity scores and plot the cons similarity vs pros similarity for a specific company 

```python
%matplotlib inline
import pandas as pd # pandas
import matplotlib.pyplot as plt # module for plotting 
import getopt
import re
import string
import nltk
import linecache
import math
import sys
import heapq
from nltk.stem.porter import *
from pylab import rcParams
rcParams['figure.figsize'] = 17, 10

#functions for searching the files

#define/declare global variables
dic_term = {}

dict_file = ""
posting_file = ""
query_file = ""
query_result_file = ""

#create a stemmer for query processing
stemmer = PorterStemmer()

#the number of results needed to return
k_value = 51

#process the query: tokenize and stem
def query_process(query):

    #split the free text query
    query_word = re.split("([ ()])", query)
    
    #remove the empty word and space
    query_word = filter(lambda x: x != "" and x != " ", query_word)
    
    index = 0
    while index < len(query_word):
        query_word[index] = stemmer.stem_word(query_word[index].lower())
        index = index + 1  
    
    return query_word
    

#read the dic from file
def read_dic():
    dic = open(dict_file)
    for line in dic:
        array = line.split()
        term = array[0]
        dic_term[term] = [int(x) for x in array[1:]]
    dic.close()


#compute the cosinescore
def cosine_score(query):
    #get the total doc num from 1 row of posting
    total_doc = eval(linecache.getline(posting_file, 1))

    #score
    doc_scores = {}
    #query tf score
    query_tf = {}
    #weighted score
    #w1^2+w2^2+w3^2+...
    sum_squre = 0

    for term in query:
        if query_tf.has_key(term):
            query_tf[term] += 1
        else:
            query_tf[term] = 1
    for term in query:

        #the dictionary contains the term
        if dic_term.has_key(term):
            
            term_df = dic_term[term][0]
            term_idf = math.log(float(total_doc)/term_df,10)
            
            term_q_tf = math.log(query_tf[term],10)

            term_score = (1 + term_q_tf) * term_idf
            sum_squre += math.pow(term_score, 2)

            #get the posting's row contain this term
            post_list = eval(linecache.getline(posting_file, dic_term[term][1]))

            #compute the score on each document having the term
            #score=(q-tf)*(d-idf)*(d-tf)
            for doc_Id, d_tf in post_list:
                
                if doc_scores.has_key(doc_Id):
                    doc_scores[doc_Id] += term_score * (1 + math.log(d_tf,10))    
                else:
                    doc_scores[doc_Id] = term_score * (1 + math.log(d_tf,10))
                    
    #no terms in found
    if sum_squre == 0: return []

    # Normalize the scores
    norm_const = 1/math.sqrt(sum_squre) * 1/float(len(query))

    for doc_id in doc_scores.keys():
        if norm_const != 0:
            doc_scores[doc_id] = doc_scores[doc_id]*norm_const
       
    # return the top k scores   
    result =  heapq.nlargest(k_value, doc_scores, key = lambda k: doc_scores[k]) 

    #convert it into string
    for k in range(0, len(result)):
        result[k] = (str(result[k]),doc_scores[result[k]])
    
    return result
        
#solve all the queries
def solve():
    
    #load dic
    read_dic()
    
    target = open(query_file)
    
    query_result = []
    #process
    query = ""
    
    for q in target:
        query += q
    
        
    query = query.strip()
    #print query
    if(query==""):
        #print "No comments"
        return []
        
    res = cosine_score(query_process(query))
    return res

def plot(name):
    pro_dict = {}
    con_dict = {}
    dir_ = ["Comments_pros/", "Comments_cons/"] 
    index = 0

    for ele in dir_:
            global dict_file
            global posting_file
            global query_file
            dict_file = "dic"+ele.split("/")[0]
            posting_file = "posting"+ele.split("/")[0]
            query_file = ele+str(name)
            k = solve()
            for ele in k:
                if index ==0:
                    pro_dict[ele[0]] = ele[1]
                else:
                    con_dict[ele[0]] = ele[1]
            index += 1

    names = sorted(pro_dict.keys())
    try:
        names.remove(name)
    except:
        name = ""
    x_cons = []
    y_pros = []
    for key in sorted(con_dict.keys()):
        if key!=name:
            x_cons.append(con_dict[key])

    for key in sorted(pro_dict.keys()):
        if key!=name:
            y_pros.append(pro_dict[key])

    plt.scatter(
    x_cons, y_pros, marker = 'o',
    cmap = plt.get_cmap('Spectral'))
    for label, x, y in zip(names, x_cons, y_pros):
        plt.annotate(label, xy = (x, y), xytext = (-1, 1),textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.1', fc = 'yellow', alpha = 1),arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
    plt.xlabel("Cons Similarity Score")
    plt.ylabel("Pros Similarity Score")
    plt.title("Pros and Cons")
    plt.show()
            
```

Comments: The top 50 similarities are computed as the tf-idf score of the input company comment with all the other companies.
